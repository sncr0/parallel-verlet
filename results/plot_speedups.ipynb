```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
data = np.loadtxt('data/electrostatic_benchmark_results.csv', delimiter=',')
fig, ax = plt.subplots()
for row in data:
    ax.plot([1, 2, 4, 8, 12, 16, 32, 64, 128, 256], row[1]/row[2:], '-', label = f"System size: {int(row[0])}")
ax.legend()
ax.axhline(1, color='k', linestyle='--')
ax.set_xlabel('Number of cores')
ax.set_ylabel('Relative Speedup')
ax.set_xscale('log', base=2)
ax.xaxis.set_major_formatter(FuncFormatter(lambda x, _: f"{int(x)}"))

fig.savefig('../report/images/electrostatic_scaling.png', dpi=300)
```

```python
def count_iterations(num_particles):
    total_iterations = 0
    for particle1_index in range(num_particles):
        for particle2_index in range(particle1_index + 1, num_particles):
            print(f"Calculating interaction between particle {particle1_index} and particle {particle2_index}")
            total_iterations += 1
    return total_iterations

# Example usage
num_particles = 10  # Adjust this value as needed
iterations = count_iterations(num_particles)
print(f"Total iterations for {num_particles} particles: {iterations}")
```
