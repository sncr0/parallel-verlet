```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt('data/electrostatic_benchmark_results.csv', delimiter=',')
fig, ax = plt.subplots()
for row in data:
    ax.plot([1, 2, 4, 8, 12, 16, 32, 64, 128, 256], row[2]/row[2:], '-', label = f"System size: {int(row[0])}")
ax.legend()
ax.axhline(1, color='k', linestyle='--')
fig.savefig('../report/images/electrostatic_scaling.png', dpi=300)
```
